```python
import mne
import matplotlib.pyplot as plt
import numpy as np
```

```python
edf_file = "D:/M.Tech/3rd Sem/trials/Dataset comparision/Siena Scalp EEG Database/PN00-1.edf"
raw = mne.io.read_raw_edf(edf_file, preload=True)
```

```python
print("\nOriginal Channel Names:")
print(raw.ch_names)
```

```python
rename_dict = {ch: ch.replace('EEG ', '') for ch in raw.ch_names}
raw.rename_channels(rename_dict)
```

```python
print(raw.ch_names)
```

```python
eeg_channels = [
    'Fp1','Fp2',
    'F7','F3','Fz','F4','F8',
    'T3','C3','Cz','C4','T4',
    'T5','P3','Pz','P4','T6',
    'O1','O2'
]
raw.pick(eeg_channels)

print(raw.ch_names)
```

```python
montage = mne.channels.make_standard_montage("standard_1020")

raw.set_montage(montage)
```

```python
raw.set_eeg_reference('average')
```

```python
raw.plot_sensors(
    kind='topomap',
    show_names=True
)

plt.show()
```
